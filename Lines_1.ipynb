```python
def getWprize(s, startIndex, curmaxLen, maxLen):
    if startIndex == len(s) - 1:
        return max(curmaxLen, maxLen)
    if (list(s)[startIndex] == 'W' and list(s)[startIndex + 1] =='W'):
        return getWprize(s, startIndex + 1, curmaxLen + 1, maxLen)
    else:
        return getWprize(s, startIndex + 1, 1, max(curmaxLen, maxLen))

def getAprize(s, startIndex, curmaxLen, maxLen):
    if startIndex == len(s) - 1:
        return max(curmaxLen, maxLen)
    if ((list(s)[startIndex]=='A' or list(s)[startIndex]=='W') and (list(s)[startIndex+1]=='A' or list(s)[startIndex+1]=='W')) :
        return getAprize(s, startIndex + 1, curmaxLen + 1, maxLen)
    else:
        return getAprize(s, startIndex + 1, 1, max(curmaxLen, maxLen))
    
def getBprize(s, startIndex, curmaxLen, maxLen):
    if startIndex == len(s) - 1:
        return max(curmaxLen, maxLen)
    if ((list(s)[startIndex]=='B' or list(s)[startIndex]=='W') and (list(s)[startIndex+1]=='B' or list(s)[startIndex+1]=='W')) :
        return getBprize(s, startIndex + 1, curmaxLen + 1, maxLen)
    else:
        return getBprize(s, startIndex + 1, 1, max(curmaxLen, maxLen))
    
def getCprize(s, startIndex, curmaxLen, maxLen):
    if startIndex == len(s) - 1:
        return max(curmaxLen, maxLen)
    if ((list(s)[startIndex]=='C' or list(s)[startIndex]=='W') and (list(s)[startIndex+1]=='C' or list(s)[startIndex+1]=='W')) :
        return getCprize(s, startIndex + 1, curmaxLen + 1, maxLen)
    else:
        return getCprize(s, startIndex + 1, 1, max(curmaxLen, maxLen))
    
def getDprize(s, startIndex, curmaxLen, maxLen):
    if startIndex == len(s) - 1:
        return max(curmaxLen, maxLen)
    if ((list(s)[startIndex]=='D' or list(s)[startIndex]=='W') and (list(s)[startIndex+1]=='D' or list(s)[startIndex+1]=='W')) :
        return getDprize(s, startIndex + 1, curmaxLen + 1, maxLen)
    else:
        return getDprize(s, startIndex + 1, 1, max(curmaxLen, maxLen))
    
def getEprize(s, startIndex, curmaxLen, maxLen):
    if startIndex == len(s) - 1:
        return max(curmaxLen, maxLen)
    if ((list(s)[startIndex]=='E' or list(s)[startIndex]=='W') and (list(s)[startIndex+1]=='E' or list(s)[startIndex+1]=='W')) :
        return getEprize(s, startIndex + 1, curmaxLen + 1, maxLen)
    else:
        return getEprize(s, startIndex + 1, 1, max(curmaxLen, maxLen))

def getFprize(s, startIndex, curmaxLen, maxLen):
    if startIndex == len(s) - 1:
        return max(curmaxLen, maxLen)
    if ((list(s)[startIndex]=='F' or list(s)[startIndex]=='W') and (list(s)[startIndex+1]=='F' or list(s)[startIndex+1]=='W')) :
        return getFprize(s, startIndex + 1, curmaxLen + 1, maxLen)
    else:
        return getFprize(s, startIndex + 1, 1, max(curmaxLen, maxLen))
```

```python
def getprize(s):
    nowprize = max(Wprize[getWprize(s,0,1,1)], Aprize[getAprize(s,0,1,1)], Bprize[getBprize(s,0,1,1)], Cprize[getCprize(s,0,1,1)], Dprize[getDprize(s,0,1,1)], Eprize[getEprize(s,0,1,1)], Fprize[getFprize(s,0,1,1)])
    return nowprize
```
